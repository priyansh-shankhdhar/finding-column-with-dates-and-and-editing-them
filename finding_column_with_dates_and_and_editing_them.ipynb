## **Creating a function to detect the columns consisting date**

Here we, at first saperate all values in colums by=>(  , space , _ , - , \ , / , st , nd ) and make them as a list, thus now because we have saprated them wilth almost all happening characters in date format so here month , year and date will get saperated.
thus if a column consist dates then it will have at least 3 values of date, month and year . Furthermore, the avg length of values will be less than 5 because we have not saperated time consisting ':'. 

```
#this function tells wether the specific column has dates or not
def get_date_columns(df):# function start 
  df.dropna(inplace=True)# first we filter the values by dropping NaN values.
  for i in df.columns:#this ffor loop is for saperating the values in a column
    df[i]=df[i].astype(str).apply(lambda x: list(re.split(',| |_|-|/|\|st|th|nd',x)))
  cols=[]
  for c in df.columns:#this for loop is for detecting the avrage length of saperated values in diffrent columnsand if length is > =3 and <=5 it is a date
    m=[]
    for i in df[c]:
      m.append(len(i))
    avg=(sum(m)/len(m))
    if avg >= 3 and avg <=5:
      cols.append(c)
  return cols # we return the list of columns with date

##############################################################
# HERE IS THE MAIN FUNCTION FOR RETURNING THE OUTPUT WE WANT #
##############################################################

def filtered(df):
  orignal=df.copy() #because while detecting the columns our orignal dataframe will be distorted thus we copy it in variable orignal
  j=get_date_columns(df)# we call the get_date_columns funtion.
  #print(j)#we print the columns with date 
  for i in j:# here we convert the dates in data frame to date-time format
    orignal[i] = pd.to_datetime(orignal[i],infer_datetime_format=True)# fromat("%Y-%m-%d")
  # Here we are subtracting columns like date1-date2 and Renaming them
  first=orignal[j[0]]#we have to create first date column, because it will change when we subtract another date from it.
  for i in range(len(j)):#here we run the loop which will act only on columns consisting date
    if len(j)==1:# for a signle column containing date we have to do nothing
      break
    elif i+1<len(j):#we want the last column case to be diffrent from others as it will subtract from first date column
      orignal[j[i]]=orignal[j[i]]-orignal[j[i+1]]#subtracting columns
      orignal.rename(columns={j[i]:str(j[i])+' - '+str(j[i+1])},inplace=True)#renaming columns
    else:
      orignal[j[i]]=orignal[j[i]]-first#subtracting the last column with first column her name of first column has changed
      orignal.rename(columns={j[i]:str(j[i])+' - '+str(j[0])},inplace=True)#renaming thw last column
  return orignal
```

# Creating the funtion to drop columns

**we know that corelation is between two colums and it can be with itself also, thus we corelated each columns containing numarical value that is integers and floats and calculated their pearson coeff., now in order to drop few columns we created an array having no. of corelations of a specific colums that are  greater that 0.85 . and dropped the columns haviing corelation coeff greater than 0.85 majority of times**

eg. colums= {
total_rooms,total_bedrooms,population,households
} all of them were havong a count of 4 ,
that means in 4 corelations they were having pearson coeff. > 0.85

```
def drop_cols(df):  
  data=df.select_dtypes(include=['int64','float64','int32','float32'])# selecting  colums with numarical value 
  cols=data.columns
  l=len(cols)
  dic={}
  for i in cols:#in this loop we fill the dictionary with value of corelation coeff of each column with each other
    dic.update({i:[]})
    for j in cols:
      corr, _ = pearsonr(data[i], data[j])
      dic[i].append(corr)
  counts=[]
  for i in dic:#in this loop we count no. of times corealtion coeff. of a a column is greater than 0.85
    j=dic[i]
    c=0
    for i in j:
      if i >= 0.85:
        c+=1
    counts.append(c)
  f=0
  v=0#counter
  for i in counts:#in this loop we drop the colums with high degree of corelation coeff. i.e; near the half no.
    if i+1 >= len(counts)/2:
      v+=1
      df.drop(columns=[cols[f]],inplace=True)
      print('{} column is dropped'.format(cols[f]))
    f+=1
  if v==0:#if no column is dropped drop the column with highest degree of corelation coeff
    f=np.argmax(counts)
    df.drop(columns=[cols[f]],inplace=True)

  return df

drop_cols(df)
```
